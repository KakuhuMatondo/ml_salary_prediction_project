```python
def shorten_categories(categories, cutoff):
    categorical_map={}
    for i in range( len(categories)):
        if categories.values[i]>=cutoff:
            categorical_map[categories.index[i]]=categories.index[i]
        else:
            categorical_map[categories.index[i]]='Other'
    return categorical_map
```

```python
country_map=shorten_categories(data.Country.value_counts(),400)
data['Country']=data["Country"].map(country_map)
data[data["Country"!='Other']]
data.Country.value_counts()
```

```python
data['YearsCodePro'].unique()
```

```python
def clean_experience(x):
    if x=="More than 50 years":
        return 50
    if x=="Less than 1 year":
        return 0.5
    return float(x)

data['YearsCodePro']=data['YearsCodePro'].apply(clean_experience)
        
```

```python
data['EdLevel'].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if  'Master’s degree' in x:
        return 'Master’s degree'
    if  'Professional degree' in x:
        return 'Post Grad'
    return 'Less than a Bachelors'

data['EdLevel']=data['EdLevel'].apply(clean_education)
```

```python
data['EdLevel'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_education=LabelEncoder()
data['EdLevel']=le_education.fit_transform(data['EdLevel'])
data['EdLevel'].unique()
```

```python
le_country=LabelEncoder()
data['Country']=le_country.fit_transform(data['Country'])
data['Country'].unique()
```

```python
X=data.drop('Salary',axis=1).astype(float)
y=data['Salary']
```

```python
X.info()
```

```python
from sklearn.linear_model import LinearRegression
linearmodel=LinearRegression()
linearmodel.fit(X,y.values)
```

```python
y_pred=linearmodel.predict(X)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
error=np.sqrt(mean_squared_error(y,y_pred))
```

```python
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_treemodel=DecisionTreeRegressor(random_state=0)
dec_treemodel.fit(X,y.values)
```

```python
y_predd=dec_treemodel.predict(X)
```

```python
error=np.sqrt(mean_squared_error(y,y_predd))
error
```

```python
from sklearn.ensemble import RandomForestRegressor
randformodel=RandomForestRegressor(random_state=0)
randformodel.fit(X,y.values)
```

```python
y_predrf=randformodel.predict(X)
```

```python
error=np.sqrt(mean_squared_error(y,y_predrf))
error
```

```python
from sklearn.model_selection import GridSearchCV
max_depth=[None,2,4,6,8,10,12]
parameters={'max_depth':max_depth}

regressor=DecisionTreeRegressor(random_state=0)
gs=GridSearchCV(regressor,parameters,scoring='neg_mean_squared_error')
gs.fit(X,y.values)
```

```python
regressor=gs.best_estimator_
regressor.fit(X,y.values)
y_predregressor=regressor.predict(X)
error=np.sqrt(mean_squared_error(y,y_predregressor))
error
```
